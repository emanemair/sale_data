```python
import pandas as pd 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv("codecamp/sales_data.csv" , parse_dates=["Date"])
```

```python
data["Unit_Cost"].mean() 
```

```python
data["Unit_Cost"].median()
```

```python
data["Unit_Cost"].plot(kind="box" , vert = False , figsize=(14,6))
```

```python
data["Unit_Cost"].plot(kind="density" , figsize=(14,6)) 
```

```python
ax = data["Unit_Cost"].plot(kind="density" , figsize=(14,6)) 
ax.axvline(data["Unit_Cost"].mean(), color='red') 
ax.axvline(data["Unit_Cost"].median(),color='green')

ax.text(data["Unit_Cost"].mean(), ax.get_ylim()[1]*0.9, 'Mean', color='grey')
ax.text(data["Unit_Cost"].median(), ax.get_ylim()[1]*0.9, 'Median', color='grey')
ax.set_title('Density Plot of Unit Cost')
```

```python
ax = data['Unit_Cost'].plot(kind='hist', figsize=(14,6))
ax.set_ylabel('Number of Sales')
ax.set_xlabel('dollars')
```
